2. 언어 분야에서 인공지능의 활용

 언어 분야에서 인공지능의 활용은 크게 두 가지로 구분할 수 있다.첫 번째는 언어 학습이고 두 번째는 언어 번역이다.

 **언어 학습 분야**
> - 인공지능형 언어학습용 1세대 로봇 : 유진로보틱스의 아이로비, 혼다의 아시모(ASIMO), 한국과학기술연구원 지능 로봇 사업단의 잉키 로봇

 > ![아이로비](https://imgnews.pstatic.net/image/003/2016/01/21/NISI20160121_0011265912_web_99_20160121085906.jpg?type=w647)
> -  2000년대 : 자연언어처리(NLP) 기술이 발전되고, 이 기술을 언어학습에  적용하는 다양한 시도가 이루어짐
> -  2010년대 : 인공지능형 로봇 언어학습용 시스템 등장, Dave ESL ChatBot는 2010년도 초, TYCHE, MUSIO, Sophia는 2016년도, 최근 Genie Tutor 원천기술을 이용한 시스템

 **언어 번역 분야**
> - 1950년대 : 기계 번역(자동 번역의 시작), 문맥 대신 단어나 구를 통해 번역하였기에 앞뒤가 맞지 않거나 비논리적인 번역이 많음

 > - **구글 번역** : 2006년 등장, 2016년에 구글 번역이 신경 기계 언어 엔진인 구글 신경 기계 번역(GNMT)으로 전환한다고 발표하였으며, 조각 단위가 아닌 한 번에 전체 문장을 번역. 신경망 번역 방식.
>> 인공지능 **머신러닝** 학습 방법 적용 : 신경망 번역이 적용된 번역 모델은 먼저 인터넷 상에 떠돌아다니고 있는 인간이 미리 번역해 놓은 콘텐츠를 웹 크롤링을 활용해 수집했다. 이를 바탕으로 번역 모델은 초기 번역 시스템을 구축하고, 구글의 개발자에게 해당 언어의 번역을 제대로 수행하고 있는지 일일이 확인 받았다(감독학습). 지도 학습이 끝난 후 번역 모델은 인간의 도움 없이 스스로 특정 언어 데이터를 수집하고 이를 바탕으로 스스로의 성능을 강화했다(비감독학습).
> - **파파고** : 네이버가 무료로 제공하는 인공지능 번역. 파파고 역시 구글 번역과 비슷하게 NMT 기술을 활용함. NMT는 Neural Machine Translation(인공신경망 기반 기계번역)의 약어로 파파고의 NMT 기술은 입력 문장을 문장벡터로 변환하는 신경망(encoder)과 문장벡터에서 번역하는 언어의 문장을 생성하는 신경망(decoder)를 대규모의 병렬 코퍼스부터 자동으로 학습.
> ![파파고](http://newsroom.etomato.com/userfiles/n(13).png)

언어 번역 분야에서 인공지능의 학습 방법이 적용되는 사례가 많다. 자동 번역이 등장한 이래로 번역 분야는 인공지능을 기반으로 급성장하였으나 인공지능을 통한 번역이 아직까지 인간의 직접 번역을 대신할 수 없다는 의견과 몇 년이래로 완전히 번역이 가능하다는 의견이 대립 중이다.

3. 음성 분야에서 인공지능의 활용
- 음성 인식 : 기계와 인간의 인터페이스
> 스마트폰 기반 인공지능 개인 비서 서비스
 - 애플의 **시리** : Siri(시리)는 iO에스용 개인 단말 응용 소프트웨어이며, 질문에 답변하고, 권고하며, 동작을 실행하는 자연어 처리를 이용.
 - 구글의 **어시스턴트** : 구글이 개발하고 2016년 5월 자체 개발자 콘퍼런스에서 발표한 인공지능 비서이다. 구글 나우와 달리 구글 어시스턴트는 양방향 대화에 참여할 수 있음
 - 삼성의 **빅스비** : 삼성전자가 자체 개발한 음성 인식 플랫폼으로, 2017년 3월 29일 공개된 삼성전자의 갤럭시 S8에 탑재된 인공지능(AI) 가상 비서. 빅스비는 지능형 검색엔진인 울프럼 알파를 적용하여 자연스러운 대화형의 인터페이스를 지원하며 인공지능 딥 러닝 기능을 탑재.
 
 그 외 인공지능을 활용한 음성 인식은 다양한 분야와 함께 사용된다. *언어 번역뿐만 아니라 사물인터넷, 자율 주행 등의 분야*에서 음성 인식 기능이 사용된다. 그 예로 애플은 인공지능 기업을 인수하여 시리와 연동이 가능한 자율주행 자동차를 개발하고 있으며 삼성은 빅스비를 가전제품에 결합한 사물인터넷을 개발 중이다.

4. 이미지 분야에서 인공지능의 활용

- 이미지 분석 작동 원리 : 이미지를 여러 패턴으로 분류해서, 패턴형 데이터를 학습하여 새로운 이미지가 주어질 때 이것이 무엇인지 정확하게 알아내는 원리. 주로 딥러닝 혹은 신경망네트워크가 사용됨.
> - 증강현실 관련 사례 : 구글은 딥러닝 기술을 광학 현미경에 적용해 조직검사시 암세포의 존재를 감지하여 알려주는 증강현실 현미경을 개발. 암세포를 감지한 후 더 쉽게 볼 수 있도록 증강현실 기반의 이미지에 표시를 해준 후 딥러닝 알고리즘이 샘플의 형체를 인식하고, 빛으로 증강현실 기반 디스플레이에 투영.
 - 상품 관련 사례 : 상품 관련 사진을 찍거나 검색해서 조회하면 그와 관련된 리뷰, 연관 상품 등을 모두 알 수 있게하는 어플리케이션이 많아지고 있음.
 - 범죄, 위조품 관련 사례 : 지문의 패턴을 구분하였다가 분석하여 범죄자를 조회하는데 사용하거나 명품의 데이터를 기반으로 위조품의 여부를 구분하는데도 사용된다.
 - AI 카메라 : 순간적으로 여러 장의 이미지를 촬영해 합성하고 AI(인공 지능) 기술로 최적화하는 방식. 2020년 2월에 출시된 삼성전자의 갤럭시 20시리즈의 카메라에도 AI가 적용되었다. 여러 개의 카메라 렌즈를 한번에 사용해 라이브 포커스, 광각 등 다양한 모드로 사진과 동영상을 촬영하고, 해당 순간을 가장 잘 보여줄 수 있는 장면을 AI 기반으로 추천하는 식이다. 
 - 그 외: Google Cloud의 Vision API는 이미지에 라벨을 할당하고 사전 정의된 수백만 개의 카테고리로 빠르게 분류할 수 있다. 객체와 얼굴을 감지하고 인쇄 및 필기 입력 텍스트를 읽으며 유용한 메타데이터를 이미지 카탈로그에 구축한다.

5. 자율주행 분야에서 인공지능의 활용

- 현재까지 자율주행 자동차 관련 인공지능은 주로 구글에서 주도해왔으나 점차 자동차 회사들이 자체적으로 인공지능을 활용한 자율 주행 기술을 개발중이다. 폭스바겐이나 도요타 같은 회사들은 2016년부터 딥러닝을 기반으로 한 기술들을 개발중이며 그 외 자율주행 기술 스타트업 회사들 역시 딥러닝과 규칙 기반 방식으로 자율주행에 필요한 기술들을 구현하고 있다. 
- 하나의 사례로 현대, 기아차는 운전자의 주행성향에 맞는 부분 자율주행을 구현하는 기술을 개발했다. 기술이 작동하는 원리는 다음과 같다
> 1. 전방카메라, 레이더 등의 센서가 다양한 운전상황에서 발생되는 정보를 지속적으로 수집
 2. ADAS의 두뇌격인 제어컴퓨터로 송신. 
 3. 제어컴퓨터가 입력된 정보로부터 운전자의 주행습관을 추출해 종합적인 주행성향을 파악. 
 4.이 과정에서 인공지능 기술 중 하나인 머신러닝 알고리즘이 적용.

6. 결론

- 예전부터 실생활에 적용되고 있는 인공지능이 많다는 것은 알고 있었다. 그러나 이제는 인공지능이 단순히 미래의 기술인 것을 넘어 이미 현재 사회에 사라져서는 안될 필수 기술이 되어버렸다는 것을 깨달았다. 아직 인공지능 기술이 완전하게 개발된 것이 아니며 계속해서 성장중이라는 것을 감안하면 앞으로의 사회가 어디까지 바뀔지 쉽게 예측이 가지 않는다.
- 마크다운을 처음 써봐서 실수가 많았다. 중간에 이미지 링크가 여러번 들어가서 오류가 나기도 했고 단락을 나누는 것도 기존에 사용하던 문서 편집 프로그램과 달라서 어려운 점이 있었다. 그러나 점차 써가면서 알게 되겠지만 기존의 프로그램보다 편리한 점도 있을 것 같다는 생각이 지배적이다.
